```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import BatchNormalization, Dense,Dropout, Flatten
from keras.callbacks import EarlyStopping

from keras.datasets import fashion_mnist
from keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Reshaping the x_train,x_test because it is gray scale

reshape(-1, 28, 28, 1) adds an extra “color channel” dimension, which CNN layers (like Conv2D) need to process images.

Grayscale → 1 channel

RGB image → 3 channels

```python
# (samples, height, width, channels)

x_train=x_train.reshape(-1, 28, 28,1)
x_test=x_test.reshape(-1, 28, 28,1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Scaling the X_train and x_test

```python
x_train=x_train/255.0

x_test=x_test/255.0
```

y

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
opti = ['SGD','Adagrad','Adadelta','RMSprop','Adam']
```

```python
def optimizer (opti):
    for i in opti:
        print(f"Blow result is for {i}")
        model=Sequential()
        model.add(Flatten(input_shape=(28,28,1)))
        model.add(Dense(128,activation='relu'))
        model.add(Dense(64,activation='relu'))
        model.add(Dense(10,activation='softmax'))
        model.summary()
        model.compile(optimizer=i,loss='categorical_crossentropy',metrics=['accuracy'])
        model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=64,epochs=5)
    return(model)    
```

```python
optimizer(opti)
```
